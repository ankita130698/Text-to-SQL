```python
from transformers import pipeline

# Load pre-trained BERT model for sentiment analysis
sentiment_classifier = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')

# Sample text
texts = [
    "I love this product! It works wonders.",
    "I'm very disappointed with the service.",
    "This is the best day of my life!",
    "I'm so frustrated with the delays."
]

# Detect sentiments
for text in texts:
    sentiment = sentiment_classifier(text)
    print(f"Text: {text}\nSentiment: {sentiment}\n")
```

The outputs of the sentiment analysis model provide insight into how text can be classified into different sentiment categories. Here, the nlptown/bert-base-multilingual-uncased-sentiment model was used, which classifies text into sentiment categories ranging from 1 to 5 stars. These categories can be interpreted as follows:

1 star: Very negative sentiment

2 stars: Negative sentiment

3 stars: Neutral sentiment

4 stars: Positive sentiment

5 stars: Very positive sentiment

**Text**: "I love this product! It works wonders."

**Sentiment**: [{'label': '5 stars', 'score': 0.9001321792602539}]

**Explanation**: The model classified this text as having a very positive sentiment (5 stars) with a high confidence score of 0.90. The use of positive words like "love" and "wonders" strongly indicates positive emotion.

```python
from transformers import pipeline

# Load pre-trained BERT model for sentiment analysis
sentiment_classifier = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')

# Sample text
texts = [
    "I love this product! It works wonders.",
    "I'm very disappointed with the service.",
    "This is the best day of my life!",
    "I'm so frustrated with the delays.",
    "Oh great, another delay. Just what I needed.",
    "Fantastic! Another bug in the system."
]

# Function to detect sarcasm
def detect_sarcasm(text):
    sarcastic_indicators = [
        "Oh great",
        "Just what I needed",
        "Fantastic",
        "Yeah right",
        "As if"
    ]
    for indicator in sarcastic_indicators:
        if indicator.lower() in text.lower():
            return True
    return False

# Detect sentiments and sarcasm
for text in texts:
    sentiment = sentiment_classifier(text)
    is_sarcastic = detect_sarcasm(text)
    sarcasm_note = " (Sarcasm detected)" if is_sarcastic else ""
    print(f"Text: {text}\nSentiment: {sentiment}{sarcasm_note}\n")
```
